```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('dataset/worldometer_data.csv')  
df = pd.read_csv('dataset/covid_19_clean_complete.csv')

# Convert date column
df['Date'] = pd.to_datetime(df['Date'])

# Select countries to analyze
countries = ['US', 'India', 'Brazil', 'Russia', 'South Africa']

# Filter
df_filtered = df[df['Country/Region'].isin(countries)]

# Plot
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df_filtered[df_filtered['Country/Region'] == country]
    plt.plot(country_data['Date'], country_data['Recovered'], label=country)

plt.title('Cumulative COVID-19 Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```
